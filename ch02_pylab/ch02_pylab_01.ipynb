```python
import math
import pylab
xmin, xmax = -2. * math.pi, 2. * math.pi
n = 50
x = [0.] * n
y = [0.] * n
dx = (xmax - xmin)/(n-1)
for i in range(n):
    xpt=xmin + i*dx
    x[i]=xpt
    y[i]=math.sin(xpt)**2
    
pylab.plot(x,y)
pylab.show()
```

## axis labels

```python
t= pylab.linspace(0., 0.1, 1000)
Vp_uk, Vp_us = 230, 110
f_uk, f_us = 50, 60
V_uk = Vp_uk * pylab.sin(2 * pylab.pi * f_uk * t)
V_us = Vp_us * pylab.sin(2 * pylab.pi * f_us * t)
pylab.plot(t*1000, V_uk, label = 'UK')
pylab.plot(t*1000, V_us, label = 'US')
pylab.title('A comparison of AC voltages in the UK and US')
pylab.xlabel('Time/ms', fontsize=16.)
pylab.ylabel('Voltage/V', fontsize=16.)
pylab.legend()
pylab.show()
```

## Using LaTeX in `pylab`

```python
import pylab
pylab.rc('text',usetex=True)#enable latex
x=pylab.linspace(-10,10,1001)
for n in range(1,5):
    y = x**n * pylab.sin(x)
    y /= max(y)
    pylab.plot(x,y,label=r'$x^{}\sin x$'.format(n))
pylab.legend(loc='lower center')
pylab.show()
# pylab.savefig('sinnx.pdf',format='pdf')# save to pdf file
```
